1、层次化索引

层次化索引是pandas的一项重要功能, 他是我们在一个轴上可以有多个索引

Series 的层次化索引

```python
s = pd.Series(np.arange(1,10),index = [['a','a','a','b','b','c','c','d','d'],[1,2,3,1,2,3,1,2,3]])
s
```

```python
s.index
```

```python
s['a']
```

```python
s['a':'c']
```

```python
s[:,1]
#两层索引用逗号隔开
```

```python
s['c',3]
```

通过unstack()将上面的Series变成DataFrame,  stack()进行相反操作

```python
s.unstack()
```

```python
s.unstack().stack()
```

离散化处理 可以用cut()函数,
pd.cut(
    x,#需要离散化的数据
    bins,#分组的依据
    right=True,#每个区间内是否包含右端点
    labels=None,#给区间命名
    retbins=False,
    precision=3,
    include_lowest=False,#每个区间内是否包含左端点
    duplicates='raise',

```python
df = pd.read_json('dbookmini.json')
```

按照评分分级

```python
df['level'] = pd.cut(df['score'],[0,3,5,7,9,10])
```

```python
df['level'] = pd.cut(df['score'], [0,3,5,7,9,10], labels = ['E','D','C','B','A'])
```

```python
df
```

按照投票人数看热门程度,评分看评价,筛选出投票多,分数高的

```python
bins = np.percentile(df['peapleNum'], [0,20,40,60,80,100])
#创建分组依据
```

```python
df['hotlevel'] = pd.cut(df['peapleNum'], bins, labels=['E','D','C','B','A'])
```

```python
df[:5]
```

```python
df.tail(10)
```

```python
df[(df.hotlevel=='A')&(df.level=='A')]
```

```python
df[(df.hotlevel=='E')&(df.level=='A')]
```
