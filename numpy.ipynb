## 2、ndarray的广播机制(重要),维度不对应时候自动补全

### 为缺失的维度补充1
### 假定缺失的元素已经有值填充

六,ndarray的排序

```python
n1 = np.array([2,4,1,7,5])
```

```python
def sotrn(nd):
    for i in range(nd.size):
        for j in range(i,nd.size):
            if nd[i] > nd[j]:
                nd[i], nd[j] = nd[j], nd[i]
    return nd
```

```python
sotrn(n1)
```

```python
np.sort(n1)#不改变n1本身
```

```python
n1
```

```python
n1.sort()#改变n1本身
n1
```
