六,ndarray的排序

```python
n1 = np.array([2,4,1,7,5])
```

```python
def sotrn(nd):
    for i in range(nd.size):
        for j in range(i,nd.size):
            if nd[i] > nd[j]:
                nd[i], nd[j] = nd[j], nd[i]
    return nd
```

```python
sotrn(n1)
```

```python
np.sort(n1)
```
