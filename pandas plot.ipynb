```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
date = pd.Series(np.random.randn(100),index = np.arange(100))
date = date.cumsum()#累加
date.plot()
plt.show()
```

```python
data = pd.DataFrame(np.random.randn(1000,4),index = np.arange(1000),columns=list('ABCD'))
data = data.cumsum()
display(data.head(10))
```

```python
data.plot()
plt.show()
```

```python
ax = data.plot.scatter(x="A",y='B',color='blur',label='class 1')
```
